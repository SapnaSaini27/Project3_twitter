Convert the target 4 to 1

0 --> Negative Tweet


1 --> Positive Tweet

**Stemming**

Stemming is the process of reducing a word into its root word.

```python
port_stem = PorterStemmer()

def stemming(content):

  stemmed_content = re.sub('[^a-zA-z]',' ',content)
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split()
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content = ' '.join(stemmed_content)

  return stemmed_content
```

```python
df.drop(['id','date','flag','user'],axis=1, inplace = True)
```

```python
df['stemmed_column'] = df['text'].apply(stemming)
```

```python
df.head()
```

```python
df['stemmed_column']
```

Seprating the data(tweet) and label.(target)

```python
x = df['stemmed_column'].values
```

```python
y = df['target'].values
```

```python
print(x)
```

```python
print(y)
```

Splitting the data for training and testing



```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, stratify=y,random_state = 2)
```

```python
x.shape, x_train.shape, x_test.shape
```

```python
print(x_train)
```

```python
print(x_test)
```

Converting Texual data to Numerical data

```python
vectorizer = TfidfVectorizer()

x_train = vectorizer.fit_transform(x_train)
x_test = vectorizer.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

Training the Machine Learning Model.

Logistic Regression

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(x_train, y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy score on training data
x_train_prediction = model.predict(x_train)
trainig_data_accuracy = accuracy_score(y_train, x_train_prediction)
```

```python
print("Accuracy of the training model is ", trainig_data_accuracy)
```

```python
# accuracy score on test data
x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(y_test, x_test_prediction)
```

```python
print("Accuracy of the testing model is ", testing_data_accuracy)
```

Model Accuracy 78.8 %
